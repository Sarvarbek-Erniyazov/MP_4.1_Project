```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


dfs_by_year = {
    2006: df_2006, 2007: df_2007, 2008: df_2008, 2009: df_2009, 2010: df_2010,
    2011: df_2011, 2012: df_2012, 2013: df_2013, 2014: df_2014, 2015: df_2015,
    2016: df_2016, 2017: df_2017, 2018: df_2018, 2019: df_2019, 2020: df_2020,
    2021: df_2021, 2022: df_2022, 2023: df_2023, 2024: df_2024, 2025: df_2025
}

# Combine all years for overall analysis
all_years = pd.concat(dfs_by_year.values(), ignore_index=True)


print("=== Missing Values Percentage per Column (Overall 2006–2025) ===")
missing_percent = all_years.isna().mean() * 100
print(missing_percent)

plt.figure(figsize=(12,6))
missing_percent.plot(kind='bar', color='tomato')
plt.title("Percentage of Missing Values per Column (2006–2025)")
plt.ylabel("Missing (%)")
plt.xticks(rotation=45)
plt.ylim(0,100)
plt.show()

# -------------------------------
# 3️⃣ Descriptive Statistics
# -------------------------------
print("=== Numeric Description (Overall 2006–2025) ===")
print(all_years.describe())

print("=== Object Columns Description ===")
print(all_years.describe(include='object'))

# -------------------------------
# 4️⃣ Daily Trends per Year
# -------------------------------
def plot_daily_trends(years, column='T'):
    plt.figure(figsize=(14,6))
    for year in years:
        df = dfs_by_year[year]
        plt.plot(df['Day'], df[column].fillna(method='ffill'), alpha=0.7, label=str(year))
    plt.title(f"Daily {column} Trends")
    plt.xlabel("Day of Year")
    plt.ylabel(column)
    plt.legend(bbox_to_anchor=(1.05,1))
    plt.show()

# Example: plot temperature for all years
plot_daily_trends(list(dfs_by_year.keys()), column='T')

# -------------------------------
# 5️⃣ Monthly Average Trends
# -------------------------------
def plot_monthly_trends(years, column='T'):
    plt.figure(figsize=(14,6))
    for year in years:
        df = dfs_by_year[year]
        monthly_avg = df.groupby('Month')[column].mean()
        plt.plot(monthly_avg.index, monthly_avg.values, marker='o', label=str(year))
    plt.title(f"Monthly Average {column}")
    plt.xlabel("Month")
    plt.ylabel(column)
    plt.legend(bbox_to_anchor=(1.05,1))
    plt.show()

# Example: plot monthly temperature for all years
plot_monthly_trends(list(dfs_by_year.keys()), column='T')

# -------------------------------
# 6️⃣ Boxplots for Outlier Detection
# -------------------------------
plt.figure(figsize=(12,6))
sns.boxplot(x='Month', y='T', data=all_years)
plt.title("Monthly Temperature Distribution (2006–2025)")
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(x='Month', y='PP', data=all_years)
plt.title("Monthly Precipitation Distribution (2006–2025)")
plt.show()

# -------------------------------
# 7️⃣ Correlation Analysis
# -------------------------------
corr = all_years[['T','TM','Tm','SLP','H','PP','VV','V','VM']].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix - Weather Variables (2006–2025)")
plt.show()

# -------------------------------
# 8️⃣ Annual / Monthly Comparison Table
# -------------------------------
monthly_avg_all = all_years.groupby(['Year','Month'])['T'].mean().unstack()
print("=== Monthly Average Temperature Table ===")
print(monthly_avg_all)

monthly_avg_all.plot(figsize=(14,6))
plt.title("Monthly Average Temperature (2006–2025)")
plt.ylabel("Temperature (°C)")
plt.show()
```
